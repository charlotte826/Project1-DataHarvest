```python
sar_temperatures=pd.read_csv('datasets/PRISM saratoga mean temp since 1895CSV.csv',delimiter=',')
sar_temperatures_df = pd.DataFrame(sar_temperatures)
sar_temperatures_df.columns = sar_temperatures_df.iloc[9]
sar_temperatures_df = sar_temperatures_df[10:]
sar_temperatures_df = sar_temperatures_df.reset_index(drop=True)
sar_temperatures_df = sar_temperatures_df.dropna(how='any',axis=0) 

#us_temperatures_df.to_csv('datasets/US-Temperatures-CSV2.csv', index=False) 
sar_temperatures_df['County']='Saratoga'
sar_temperatures_df['State']='New York'
sar_temperatures_df['Latitude']='43.1075'
sar_temperatures_df['Longitude']='-73.8639'
sar_temperatures_df = sar_temperatures_df[['Longitude', 'Latitude', 'Date', 'tmean (degrees F)','County','State']]
sar_temperatures_df = sar_temperatures_df.rename(columns={'tmean (degrees F)': 'tmeanF',
                                                          'Date':'year',
                                                         'Longitude':'longitude',
                                                         'Latitude':'latitude'})
sar_temperatures_df
```

```python
sar_temperatures_df['longitude'] = pd.to_numeric(sar_temperatures_df['longitude'])
sar_temperatures_df['latitude'] = pd.to_numeric(sar_temperatures_df['latitude'])
#sar_temperatures_df['tmeanF'] = pd.to_numeric(sar_temperatures_df['tmeanF'])
sar_temperatures_df["tmeanF"] = sar_temperatures_df.tmeanF.astype(float)
#sar_temperatures_df.tmeanF.dtype
#https://stackoverflow.com/questions/48094854/python-convert-object-to-float
```

```python
df = sar_temperatures_df
class AcledExplorer(object):

    def __init__(self, df):
        self._df = df
        self._heatmap = None
        self._slider = None
        initial_year = min(self._df['year'])

        title_widget = widgets.HTML(
            '<h3>Mean Temperature in Saratoga County, NY, by year (1895-2017)</h3>'
            '<h4>Data from <a href="http://prism.oregonstate.edu/">PRISM Climate Group</a></h4>'
        )
        map_figure = self._render_map(initial_year)
        controls = self._render_controls(initial_year)
        self._container = widgets.VBox([title_widget, controls, map_figure])
    def render(self):
        display(self._container)
    def _on_year_change(self, change):
        year = self._slider.value
        self._heatmap.locations = self._locations_for_year(year)
        self._total_box.value = self._total_casualties_text_for_year(year)
        return self._container
    def _render_map(self, initial_year):
        fig = gmaps.figure(map_type='HYBRID')
        self._heatmap = gmaps.heatmap_layer(
            self._locations_for_year(initial_year),
            max_intensity=100,
            point_radius=8
        )
        fig.add_layer(self._heatmap)
        return fig
    def _render_controls(self, initial_year):
        self._slider = widgets.IntSlider(
            value=initial_year,
            min=min(self._df['year']),
            max=max(self._df['year']),
            description='Year',
            continuous_update=False
        )
        self._total_box = widgets.Label(
            value=self._total_casualties_text_for_year(initial_year)
        )
        self._slider.observe(self._on_year_change, names='value')
        controls = widgets.HBox(
            [self._slider, self._total_box],
            layout={'justify_content': 'space-between'}
        )
        return controls
    def _locations_for_year(self, year):
        return self._df[['latitude', 'longitude']]
    def _total_casualties_for_year(self, year):
        return int(self._df[self._df['year'] == year]['tmeanF'])
    def _total_casualties_text_for_year(self, year):
        return '{} *F'.format(self._total_casualties_for_year(year))
AcledExplorer(df).render()
```
